## **QUES 6)Explain the difference between Lasso, Ridge, and Elastic Net regression**

ANS


Lasso, Ridge and elastic_net are  concept of regularisation that is used to make the model efficient.

LASSO

Lasso is also known as L1 penalty that is used to remove unnecessary features. It is used to select those features that are relevant in actual. It automatically reduce the coefficients to Zero .
Here the hyper parameter that is lambda focus on reducing features. As lambda increase Cost function is also increase and slope tends to 0 as well .

    Penalty:- lambda*|m|

RIDGE

It is also called L1 penalty.
Ridge is another type of regularisation that helps in reducing overfitting problem. Its main focus is on overfitting but not on removing irrelevant features.
Here , if lambda increase , Cost function is also increase and minima is shifted is also increase . Apart from that slope is decline but never become zero .

    Penalty:- lambda * m²


ELASTIC_NET

This is last regularisation type that is useful to perform both both L1 and L 2 penalty. It focus on reducing insignificant features and dealing with overfitting as well.

    Penalty:- lambda*m² + lambda * |m|


## **QUES 7) When should we use Elastic Net instead of Lasso or Ridge**.

So, If one want to remove unnecessary features alongs with reducing overfitting then he/she may use Elastic_net.

    Penalty:- Cost function + L1 + L2


## **QUES 8) What is the impact of the regularization parameter (λ) in Logistic Regression**.

ANS

Regularisation parameter (lambda)

Impact

- when lambda is **high** , it tends to impose more stronger  penalty on large weight which leads to make simpler model.
- it reduce overfitting issue.
- it also keep shifting the minima in positive manner.
- High lambda can leads to remove insignificant features.

- When lambda is **low** , it may impose weaker penalty of large weight which leads to built more complex model.
- It also leads to increase overfitting issue.
- low lambda can also impact on variance . It may increase the variance as model may be too sensitive for small data .

- When lambda is **balanced** it create optimum balance between trade off  bias and variance &  it also reduce error from the model.

## **QUES 11) What are Classification Evaluation Metrics**.

ANS


There are multiple Classification Metrics for logistic regression.

1) Confusion metrix :- This is 2 by 2 metrix where 4 areas are classified:-

- True Positive
- True negative
- False positive
- False negative

2) Accuracy metrix :- It can be calculated through formula
   
           TP +TN /TP+ TN +FP+FN

It states how many are truly predict out of all data points.

3) Precision: It states out of actual values how many are correctly predicted.


                   TP/(TP+FP )

4) Recall : It states out of predicted values how many are correctly predicted with actual values.

                  TP / TP +FN

5) F-Beta score :


           (1+B²) x (Precision * Recall) / (Precision + recall)

if FN and FP both are important then B = 1

if only FN is  important then B = 2

if only FP is important then B = 0.5

6) True Positive rate : It is also called recall and sensitive .

          TP/TP+FN

7) False positive rate :

          FP/FP + TN

8) True negative rate :

          TN /FP +TN

9)ROC- AUC (Receiver  operating characteristics - area under cover )

10) Precision accuracy recall tradeoff

## **QUES 13) Write a Python program to train a Logistic Regression model on imbalanced data and apply class weights toimprove model performance**


```python
# Generate imbalanced data
np.random.seed(42)
X = np.random.rand(500, 2) * 10
Y = (X.sum(axis=1) > 12).astype(int)  # Imbalance by shifting threshold

# Split data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)

# Train logistic regression with class weights
model = LogisticRegression(class_weight='balanced')
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)


# Print classification report
print(classification_report(Y_test, Y_pred))

# Compute and print precision, recall, and F1-score
precision = precision_score(Y_test, Y_pred)
recall = recall_score(Y_test, Y_pred)
f1 = f1_score(Y_test, Y_pred)
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')
```

## **QUES 14) Write a Python program to train Logistic Regression on the Titanic dataset, handle missing values, and evaluate performance**

```python
# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Select relevant features and target
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
df = df[features + ['Survived']]

# Handle missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Encode categorical variables
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)

# Split data
X = df.drop(columns=['Survived'])
Y = df['Survived']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)


# Train logistic regression model
model = LogisticRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)


# Print classification report
print(classification_report(Y_test, Y_pred))

# Compute and print precision, recall, and F1-score
precision = precision_score(Y_test, Y_pred)
recall = recall_score(Y_test, Y_pred)
f1 = f1_score(Y_test, Y_pred)
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')
```

## **QUES 24) Write a Python program to train Logistic Regression and find the optimal C (regularization strength) using cross-validation**


```python
x,y=make_classification(n_samples=100,n_features=5,n_redundant=2,n_informative=2,n_classes=2,random_state=1)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=1)

from sklearn.model_selection import GridSearchCV
param_grid={"C":[0.1,1,10,100,1000]}
```

```python
model=GridSearchCV(LogisticRegression(),param_grid,cv=5,scoring='accuracy')
model
```

```python
model.fit(x_train,y_train)
```

```python

# get optimum cv
#"C":[0.1,1,10,100,1000]
C_optimum=model.best_params_['C']
C_optimum
```

```python
Model=LogisticRegression(C=C_optimum)
Model
```

```python
Model.fit(x_train,y_train)
```

```python
y_pred=Model.predict(x_test)

## accuracy

print(f'accuracy_score : {accuracy_score(y_pred,y_test)}')
```

## **QUES 25 ) Write a Python program to train Logistic Regression, save the trained model using joblib, and load it again to make predictions**.

```python
import joblib

# Generate synthetic data
X, y = make_classification(n_samples=1000, n_features=20, random_state=1)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Train a Logistic Regression model
model = LogisticRegression()
model.fit(X_train, y_train) # fit model

# Save the trained model to a file
model_filename = 'logistic_regression_model.pkl'
joblib.dump(model, model_filename)

# Load the model back from the file
loaded_model = joblib.load(model_filename)

# Make predictions on the test set using the loaded model
y_pred = loaded_model.predict(X_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print("Evaluation Performance:")
print("-----------------------")
print("Accuracy:", accuracy)
print(report)
```
