## QUES 6) How do we handle categorical variables in Machine Learning? What are the common techniques ?

ANS-

Handling categorical variables is crucial in machine learning as Machine can not directly process the category data type. Here , we have to convert the qualitative data into quantitative data by data encoding. There are various techniques to do this.

1) Nominal / OHE :
One hot encoding can helps to create binary vector for each category . Binary value can be 0 and 1
And on basis of this category data is converted into numerical.


2) Lable technique: Here integer value is assign to  categorical data . Replacing the categorical variables with numerical one helps to learn the pattern easily for machine.

3) ordinal encoding: this technique helps to convert ordinal categorical data into numerical data without disrupting the hierarchy and order. It's is preferable for ordinal data .

4) Target guided ordinal encoding: This technique is useful to encode ordinal variable along with taking target variable in account.
For each category variable , it calculate mean target  variable.
And then rank the categorical variable based on their mean target variables.

Following are steps to approach Machine learning problem:-

1) Define problem : identify the goal of business and problem one wants to solve with machine learning. Problem can be Multiple that affect the business.

2) Data collection: Gather relevant data from different source that related to your problem.

3) data preprocessing: This is process of clearing the data and handling missing value , outliers and transform data into numerical form.

4) feature engineering: Selection of the relevant features that contribute to solve the problem  creating new features through feature extraction and modify existing one.

5) Choose model : Selecting most appropriate and suitable algorithm based on problem type.

6) Train and Evaluate the model :  Train the machine learning model using training set. And evaluating the model using validation set.

**SPLITTING THE MODEL**:

It can be done through train_test_split module :-


```python
from sklearn.model_selection import train_test_split
import pandas as pd

# data
data= {'Feature_1':[1,2,3,4,5,6,7],
       'Feature_2': [11,12,13,14,15,16,17],
       'target':[0,0,0,1,1,0,0]}

df=pd.DataFrame(data)
df
```

```python
## split data into feature(X) and target(Y)

X=df[['Feature_1','Feature_2']]  # features
Y=df['target']  # target


X_train , Y_test, Y_train,Y_test = train_test_split(X,Y,test_size=0.3, random_state=42)

X_train , Y_test, Y_train,Y_test
```

## QUES 16)What is an Optimizer? What are different types of optimizers? Explain each with an example.

ANS-

Optimezer is crucial element in deep learning that helps in adjust the neural network's parameters during training. The primary role of optimizer is to minimize the error and make model more efficient.These optimizer widely effect the accuracy and speed of deep learning model during training.

Types of optimizer

1)  Gradient descent
It is an iterative optimization algorithm that uses the gradient to search for optimal solution.
It updates the parameter in the direction of the negative gradient.

- Example - suppose one want to find bottom of hills . here GD is like taking small steps down the hill, following the steepest path .


2) Stochastic Gradient descent :
It is part of GD that uses a random subset of data to compute the gradient at each iteration.

   - Ex - similar to GD , but instead of looking at the whole hill, you only look at small part of it and take a step down, then you look at another small part and take another step.

3) Adam optimizer : Adaptive moment estimation is popular optimizer that adapts the learning rate of each parameter based on the magnitude of gradient.
- Example - adam like a personal trainer , it helps to take step down the hill but also adjusts your pace based on how steel the hill is.


4) RMSProp optimizer : It is an optimizer that divides the learning rate by a moving average of the squared gradient.

- Example - it provides safety net that helps you to take steps down the hill with taking safety measures .
