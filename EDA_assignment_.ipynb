##**QUES 11) Identify and remove outliers in the km_driven column using the IQR method.**

```python
import numpy as np
# calculating q1 and q3
q1=np.percentile(dt['km_driven'],25)
q3=np.percentile(dt['km_driven'],75)

# IQR
IQR=q3-q1
IQR
```

```python
# outlier
lower_bound=q1-1.5*IQR
upper_bound=q3+1.5*IQR
outlier=dt[(dt['km_driven']<lower_bound)|(dt['km_driven']>upper_bound)]
```

```python
# remove outlier
rem=dt[(dt['km_driven']>=lower_bound)&(dt['km_driven']<=upper_bound)]

print( len(outlier)) # outlier removed
print(len(rem)) # row cleaned
```
