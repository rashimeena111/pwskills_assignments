##**3.Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?**

ANS-


Dispersion states that how data points are spread around the central tendency. It measures how much individual points are spread out around Central tendency.

If the dispersion is low then data must be cluster closed to each other and if dispersion is high then data is spread more.


#1)Variance

Variance is used to measure how much individual data points are deviate from mean. It is calculated by taking average of squared  difference between each data points and mean.


  - Formula = £(Xi - μ)²/n  

Here X is is individual data points

n is no. Of data points

μ is mean

There can be two different types of variance

i) **Sample mean**- In this sample is taken from large population and then calculate the Variance.

    Formula
    £(xi-μ)²/n-1

We take n-1 in denominator as it protect the result from any biased happening. So it is play role  of unbiased estimator.


ii)**Population Variance** - In this we take whole population and then calculate variance.

    Formula = £(Xi-μ)²/n  



   Conclusion = if the variance is high then data is spread far and if variance is low then data is cluster close.


2)**Standard deviations** - It is squared root of variance. It is used to calculate average distance between each data points and mean. If standard deviations is high then data is spread out more or if it is low then data set are cluster closed .

    Formula = √variance

There are 3 types of skewness .

##**8. Discuss the conditions under which the binomial distribution is used.**
ANS-
Binomial distribution is statistics concept that refers to discrete probability of getting only two possible outcomes in experiment either fail or pass.

In simple language, it states probability of getting x result in n independent trial where each trial has probability of equal success .

 - Example - Tossing 2 coins .

Conditions under which BD is used

1) There must be only two possible outcomes.ex- Pass/fail , Good/bad , Head/tail.

2) Each outcome has equal possibility of success.

3) Each trial is independent in nature i.e., no one affect any of the outcome.

4) No. Of trial must be fixed and known.

5) trails are randomly sampled from population.No biasedness should occurs.
