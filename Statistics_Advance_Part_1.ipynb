##**QUES 3) Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution**

```python
def bernoulli_pmf(x, p):
    if x not in [0, 1]:
        return 0  # Bernoulli distribution is only defined for x in {0, 1}
    return p**x * (1 - p)**(1 - x)

# Example usage
p = 0.6  # Probability of success
print("P(X=0):", bernoulli_pmf(0, p))
print("P(X=1):", bernoulli_pmf(1, p))
```

##**QUES 6) Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution**

```python
low, high = 1, 10  # Range: Uniformly distributed between low and high (inclusive)
size = 1000  # Number of samples

# Generate discrete uniform data
data = np.random.randint(low, high + 1, size)

# Compute probability mass function (PMF)
values, counts = np.unique(data, return_counts=True)
pmf = counts / size

# Compute cumulative distribution function (CDF)
cdf = np.cumsum(pmf)
cdf
```

```python
plt.figure(figsize=(8, 5))
plt.step(values, cdf, where="post", linestyle='-', marker='o', color='b', label="CDF")
plt.xlabel("Values")
plt.ylabel("Cumulative Probability")
plt.title("(CDF) of Discrete Uniform Distribution")

plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.legend()
plt.show()
```

##**QUES 15)Implement hypothesis testing using Z-statistics for a sample dataset**

```python
from scipy import stats

# Sample data and parameters
data = np.array([12, 15, 14, 13, 16, 18, 14, 15, 13, 17])
mu_0 = 15  # Hypothesized population mean
alpha = 0.05  # Significance level

# Z-statistic and p-value
z_stat = (np.mean(data) - mu_0) / (np.std(data, ddof=1) / np.sqrt(len(data)))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Decision
print(f"Z-statistic: {z_stat:.3f}, P-value: {p_value:.3f}")
print("Reject H0" if p_value < alpha else "Fail to reject H0")
```

##**QUES 16)  Create a confidence interval for a dataset using Python and interpret the result**

```python
data = np.array([12, 13, 14, 13, 19, 16, 14, 15, 13, 17])

# Parameters
confidence_level = 0.95
alpha = 1 - confidence_level
sample_mean = np.mean(data)
sample_std_dev = np.std(data, ddof=1)  # Sample standard deviation
n = len(data)

# Z-critical value for 95% confidence
z_critical = norm.ppf(1 - alpha / 2)

# Margin of error
margin_of_error = z_critical * (sample_std_dev / np.sqrt(n))

# Confidence Interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Output the result
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
```

##**QUES 24) Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution**

```python
from scipy.stats import skew, kurtosis

# Function to calculate skewness and kurtosis
def calculate_skewness_kurtosis(data):
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)

    return data_skewness, data_kurtosis
```

```python
data = np.random.normal(0, 1, 1000)  # Mean=0, Std=1, Size=1000

# Calculate skewness and kurtosis
skewness, kurt = calculate_skewness_kurtosis(data)

# Output the results
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurt:.4f}")
```
