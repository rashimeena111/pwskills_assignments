##**QUES 15  How does Python's garbage collection system work**
Ans-
Python's garbage collection is mechanism in which python automatically frees up memory space occupied by objects that are no longer in use .

How its works:

1) Python use reference code algorithm in which it provides reffernce  count to each object and if new reffernce create then it makes increment in reference count.

2) Similarly, if reference is deleted then reference count decreases and when it goes to zero it will leads to move in garbage collection.

3) Garbage collection frees up the occupied  space to run the program smoothly and faster.

##**Ques 16 Demonstrate how to use memory profiling to check the memory usage of a small program.**

```python
import psutil
import os

def fun():
  List=[ i for i in range(100000)]
  return List

if __name__ == "__main__":
  process=psutil.Process(os.getpid())
  mem_before=process.memory_info().rss/(1024*1024)
  print(f'Memory usage before is {mem_before} MB')

  fun()

  mem_after= process.memory_info().rss/(1024*1024)
  print(f'Memory usage after is {mem_after} MB')
```

## **Ques 18 How would you implement a basic logging setup that logs to a file with rotation after 1MB.**

```python

import logging
from logging.handlers import RotatingFileHandler
```

```python
A=logging.getLogger('MyLogger')
A.setLevel(logging.DEBUG)
handler= RotatingFileHandler('namy.log', maxBytes=1_048_576, backupCount=5)
handler.setFormatter(logging.Formatter('%(asctime)s - %(levelname)s - %(message)s'))
A.addHandler(handler)


A.info('started')
A.error("Error")
```
