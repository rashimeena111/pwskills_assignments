##**Ques 9) What is the main difference between Simple and Multiple Linear Regression?**
ANS


Difference between SLR and MLR :-

- Simple linear regression focus on determining relationship between only 2 variables.Whereas in Multiple linear regression, relationship is determined among multiple variables.

- In SLM, target variable is depend on one independent variables whereas in MLR, target variable is depend on multiple variables.

- In SLM, y predicted is calculated through
          
          y= Qo + Q1X

In MLR , y predicted is calculated through

          y=Qo+ Q1x1 +Q2x2 +Q3x3.......+Qnxn.

- Both technique used Gradient descent to figure out best possible fit line.

- cost function of SLR :-
  
          (yi-y predicted)/n
Cost function of MLR:-
  
         (yi-Q1X1+Q2X2...QnXn)/n

It assign unique integer to each category.
