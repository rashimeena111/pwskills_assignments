##**QUES 15 How does NumPy handle multidimensional arrays**
ANS-

Numpy allows to create multiple array with different ways:

1) np.array() : This is simple method to create array by Converting the list or other data structures into array.

For example: A= [1,2]

         np.array(A)

2) np.asarray() : This is traditional method to create array initially when np.array() not founded.

3) np.zeros: it helps to create array with 0 elements at every place.
            
            np.zeros(4)
             OUTPUT: ([0,0,0,0])

4) np.ones() : it helps to create array with element 1 at every place .
         
         np.array(4)
         ([1,1,1,1])

5) np.random.rand() : It gives random number from input.
Ex:

               a=[2,4,6,8]
              np.random.rand(a)

6) np.arange(): it helps to create array within provided range as input.

            np.arange(1,10)
            output : ([1,2,3...9])
