##**QUES 6)Explain the working principle of a Bagging Classifier**
ANS

Bagging classifier is ensemble technique that train multiple models parallely that are  independent of each other . It takes discrete data only .

Working:-
1. Bagging classifier divide the training data into different samples with replacement to create subsets of sample. These sample are equal in size always .  This is called bootstrap  sampling also.
2. After this, on each sample model is trained accordingly.
3. Then each model makes prediction on test data.
4. Prediction from each model is Aggregate and whichever Prediction is most frequent is ultimately selected as prediction value. Ex- if Model 1,2,5 are predicted 0 and model 3,4 predicted 1 then majority prediction will be considered.

##**QUES 18) What is the role of bootstrap sampling in Bagging**

ANS

Bootstrap sampling involves randomly selecting samples from training data with replacement.

Multiple bootstrap samples are created , each of the same size as the original training data.

These diverse bootstrap are further used in model training that reduce the chances of overfitting by averaging out the errors .

##**QUES 37) Train a Stacking Classifier using Decision Trees, SVM, and Logistic Regression, and compare accuracy**

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define base models
dt = DecisionTreeClassifier()
svm = SVC()
logreg = LogisticRegression()

from sklearn.ensemble import StackingClassifier
stacking_clf = StackingClassifier(
    estimators=[('dt', dt), ('svm', svm), ('logreg', logreg)],
    final_estimator=RandomForestClassifier(n_estimators=100, random_state=42),
    stack_method='auto'
)
stacking_clf
```

```python
models = {'Decision Tree': dt, 'SVM': svm, 'Logistic Regression': logreg, 'Stacking Classifier': stacking_clf}
```

```python

accuracy_results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_results[name] = accuracy
    print(f'{name} Accuracy: {accuracy:.4f}')
```

##**QUES 40) Train a Random Forest Classifier and analyze the effect of max_depth on accuracy**

```python
data1=load_breast_cancer()
a,b=data1.data,data1.target
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.2, random_state=42)

max_depth=range(1,20)
testing_accuracy=[]
for depth in max_depth:
  clf=RandomForestClassifier(max_depth=depth,random_state=1)
  clf.fit(a_train,b_train)
  b_p=clf.predict(a_test)
  testing_accuracy.append(accuracy_score(b_p,b_test))
```

```python
import matplotlib.pyplot as plt
plt.plot(max_depth, testing_accuracy, marker='*',color='grey')
plt.xlabel("max_depth")
plt.ylabel("Testing accuracy")
plt.title("Effect of Max Depth on Accuracy")
plt.show()
```
