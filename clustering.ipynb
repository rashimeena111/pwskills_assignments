##**QUES 1)  What is unsupervised learning in the context of machine learning**

##**QUES 2)  How does K-Means clustering algorithm work**
ANS

K means clustering is a  type of  unsupervised machine learning algorithms that is use for linear problems.

1. The first step is to understand the data whether it is linear or non-linear with the help of visualization. If it is linear, then KMeans clustering can be use
2. Then determine the number of cluster you want to group your data i.e., initialize the value of K.
3. Each and every data-points are cluster in its own , so we randomly decide the centroid for each cluster .
4. Then assign each data points to the nearest cluster with the nearest centroid .
5. Then again select the centroid and assign data-points to nearest cluster with nearest centroid.
6. These steps keeps on repeating until centroid stop changing.

##**QUES 9)  How does DBSCAN identify noise points**
ANS

DBSCAN calculate the density of each data point by calculating the number of points within the specific radius .

- It categorise the data points based on threshold set earlier .
- Points with density above minimum points are considered core points
- points with density below the threshold are considered border points .
- points that are neither core nor border points are considered noise points .
- It is easy to capture by eyes with the help of Visualization tools.


##**QUES 10) Define inertia in the context of K-Means**
ANS

Inertia in KMeans also called WCSS (within cluster sum of squared distance) .

- Inertia measures the compactness of clusters. compactness means how tightly datapoints are clustered each other .
- If inertia is low the it explains the each datapoints are closer to their cluster centroids .
- It helps us to determine the optimal K .
- inertia helps use to evaluate the quality of clusters .

##**QUES 15)  Explain the term "linkage criteria" in hierarchical clustering**

ANS

Linkage criteria refers to the method used for determining the distance between clusters.
There are mainly 4 type of linkage method .

1) single linkage : Distance Between two clusters is determined  by minimum distance between any two points in clusters . It is sensitive towards outliers.

2) Complete linkage:- The distance between two clusters is determined by calculating maximum distance between any two points of clusters. It is more robust than single linkage.

3) Average linkage:- As the name states itself that the distance between two clusters is determined by taking average of distance between all points in the clusters.

4) Ward's linkage:- athe distance between two clusters is determined by the variance of points in the clusters.

##**QUES 29) Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart**


```python
from sklearn.datasets import make_blobs
x,y=make_blobs(n_samples=500,centers=4,n_features=2)
```

```python
from sklearn.metrics import silhouette_score
k_p=[2,3,4,5]
acc={}
for i in k_p:
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(x)
  y_label=kmeans.labels_
  accuracy=silhouette_score(x,y_label)
  acc[i]=accuracy
  print(f'k is {i} and accuracy score is {accuracy}')
```

```python
plt.figure(figsize=(5,4))
plt.bar(range(2,6),acc, color=['red','yellow','green','pink'])
plt.xlabel('k values')
plt.ylabel('accuracy')
plt.legend()
```
